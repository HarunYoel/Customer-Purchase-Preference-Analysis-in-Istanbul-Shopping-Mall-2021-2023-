## 2. melihat usia yang paling banyak dan rata-rata usia yang membeli produk di kategori penjualan tertinggi yaitu 'Clothing'

```python
# Filter data untuk kategori 'Clothing'
clothing_data = df[df['category'] == 'Clothing']

# Kelompokkan data berdasarkan usia dan hitung jumlah penjualan
age_sales = clothing_data.groupby('age')['total_sales'].sum()

# Hitung rata-rata penjualan per usia
average_sales_per_age = age_sales.mean()

# Temukan usia dengan penjualan tertinggi
top_age = age_sales.idxmax()
top_sales = age_sales.max()

# Cetak hasilnya
print(f"Usia yang paling banyak membeli produk Clothing: {top_age} tahun")
print(f"Total penjualan untuk usia tersebut: {top_sales}")
print(f"Rata-rata penjualan per usia: {average_sales_per_age:.2f}")
```

Kelompok usia yang paling banyak membeli produk Clothing adalah usia 27 tahun, dengan total penjualan sebesar 2.448.352,72.
dan rata-rata penjualan pada produk clotihing usia 21 tahun.
Ini menunjukkan bahwa kelompok usia ini memiliki ketertarikan atau minat yang paling besar terhadap produk-produk 'Clothing' dibandingkan dengan kelompok usia lainnya.
Dengan demikian, strategi pemasaran dapat difokuskan pada kelompok usia 21-27 tahun, mengingat mereka menunjukkan minat dan daya beli yang tinggi terhadap produk 'Clothing'.

## 6. Rata-rata pengeluaran per Transaksi untuk Setiap Metode Pembayaran (Kategori Clothing) apakah sama atau beda?

```python
# Filter data untuk kategori Clothing
clothing_data = df[df['category'] == 'Clothing']

# Hitung rata-rata pengeluaran per transaksi untuk setiap metode pembayaran
avg_spending_per_transaction = clothing_data.groupby('payment_method')['total_sales'].mean().sort_values(ascending=False)

# Visualisasi
plt.figure(figsize=(6, 4))
avg_spending_per_transaction.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Rata-rata pengeluaran per transaksi Berdasarkan Metode Pembayaran (Kategori Clothing)')
plt.xlabel('Metode Pembayaran')
plt.ylabel('Total Penjualan (Rp)')
plt.xticks(rotation=45)
plt.tight_layout()

# Tampilkan grafik
plt.show()
# Menampilkan hasil
print("Rata-rata Pengeluaran per Transaksi untuk Setiap Metode Pembayaran (Kategori Clothing):")
print(avg_spending_per_transaction)
```

Ternyata berbeda, Rata-rata pengeluaran per transaksi untuk kategori Clothing menunjukkan bahwa pelanggan yang menggunakan kartu debit memiliki rata-rata pengeluaran tertinggi sebesar Rp3.330,91, diikuti oleh pembayaran tunai dengan rata-rata Rp3.319,67, dan pembayaran dengan kartu kredit memiliki rata-rata pengeluaran terendah sebesar Rp3.272,48. Hal ini menunjukkan bahwa pelanggan yang menggunakan kartu debit dan tunai cenderung mengeluarkan lebih banyak per transaksi dibandingkan dengan mereka yang membayar menggunakan kartu kredit.

Promosi dan Diskon: Berikan promosi atau diskon khusus untuk transaksi menggunakan kartu debit dan tunai, karena pelanggan dengan metode pembayaran ini cenderung mengeluarkan lebih banyak per transaksi.

## 11.Berapa rata-rata penjualan per transaksi dan seberapa besar variasi dalam jumlah penjualan antar transaksi?

```python
# Menghitung rata-rata dan standar deviasi
mean_sales = df['total_sales'].mean()
std_sales = df['total_sales'].std()

# Membuat box plot
sns.boxplot(data=df, x='total_sales')
plt.title('Box Plot Total Penjualan')
plt.xlabel('Total Penjualan')
plt.show()

# Menampilkan hasil perhitungan
print(f"Rata-rata Penjualan: {mean_sales}")
print(f"Standar Deviasi Penjualan: {std_sales}")
```
